```python
import os,sys
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from scipy.special import erf
import pandas as pd
from astropy.table import Table, vstack
import scipy
import scipy.constants as constants

# Everything above should import with no issue, everything below needs directing
homepath = "/global/homes/c/cpopik/"
packpath = homepath+"Packages/"
resultspath=homepath+"Results/"

import camb
sys.path.append(packpath + "hmvec/")
import hmvec as hm
```

```python
table = Table.read(homepath + 'DESI_Data/DESI_LRGs_ACTD56.csv')
targetdata = table.to_pandas()

# Sets RA limits to match ACT maps
targetdata['RA'].where(targetdata['RA'] < 180, targetdata['RA'] - 360, inplace=True)

data_zs = targetdata['Z_PHOT_MEAN'].values
area = (max(targetdata.RA)-min(targetdata.RA))*(max(targetdata.DEC)-min(targetdata.DEC))

gdndz, zbins = plt.hist(data_zs, bins=np.arange(0, 1.5, 0.05), ec='w', weights=np.ones(len(data_zs))/area, density=True)[0:2]
gzs = zbins[:-1]+((zbins[1]-zbins[0])/2)
zbins = np.arange(0.4, 1.1, 0.05)
```

```python
# Create model and load up battaglia pres model
zs = zbins
ms = np.geomspace(10**11, 10**14, 200)  # That min is a min on the LRG HOD
ks = np.geomspace(1e-4, 100, 1001)
hcos = hm.HaloModel(zs,ks,ms=ms)

xmax = 30
nxs = 5000
hcos.add_battaglia_pres_profile("y",family='pres', xmax=xmax,nxs=nxs, ignore_existing=True)
```

```python
# Define and load in DESI LRG HOD model
def Nc_LRG(logM, params):
    f_ic, logM_cut, sigma = [params.get(key) for key in ['f_ic', 'logM_cut', 'sigma']]
    return (f_ic / 2) * (1- erf((logM_cut - logM)/(np.sqrt(2)*sigma)))

def Ns_LRG(logM, Nc, params):
    kappa, logM_cut, logM_1, alpha = [params.get(key) for key in ['kappa', 'logM_cut', 'logM_1', 'alpha']]
    M, M_cut, M_1 = 10**logM, 10**logM_cut, 10**logM_1
    return ((M - kappa * M_cut) / M_1)**alpha * Nc

LRG_paper_params = {
    'logM_cut': np.log10(hcos.h) + 12.89*np.ones(hcos.ms.shape),  # 12.7
    'logM_1': np.log10(hcos.h) + 14.08*np.ones(hcos.ms.shape),  # 13.6
    'sigma': 0.27*np.ones(hcos.ms.shape),  # 0.2
    'alpha': 1.20,  # 1.15
    'kappa': 0.65,  # 0.08
    'f_ic': 0.92  # .08
}

mthresh=10**LRG_paper_params['logM_cut'][0] + zs*0.
hcos.add_custom_hod(name="LRG_HOD",mthresh=mthresh, ignore_existing=True, 
                    Nc_model=Nc_LRG, Ns_model=Ns_LRG, 
                    model_params=LRG_paper_params, z_dep=False)
```

```python
logmcuts = np.array([12.89]*4+[12.78]*4+[12.89]*3+[12.68]*3) + np.log10(hcos.h)
logm1s = np.array([14.08]*4+[13.94]*4+[13.96]*3+[13.60]*3) + np.log10(hcos.h)
sigmas = np.array([0.27]*4+[0.23]*4+[0.37]*3+[0.53]*3)
alphas = np.array([1.20]*4+[1.07]*4+[0.91]*3+[0.72]*3)
kappas = np.array([0.65]*4+[0.55]*4+[0.74]*3+[0.51]*3)
fics = np.array([0.92]*4+[0.89]*4+[0.92]*3+[0.19]*3)
```

```python
# Define and load in DESI LRG HOD model
def Nc_LRG(logM, params):
    f_ic, logM_cut, sigma = [params.get(key) for key in ['f_ic', 'logM_cut', 'sigma']]
    return (f_ic / 2) * (1- erf((logM_cut - logM)/(np.sqrt(2)*sigma)))

def Ns_LRG(logM, Nc, params):
    kappa, logM_cut, logM_1, alpha = [params.get(key) for key in ['kappa', 'logM_cut', 'logM_1', 'alpha']]
    M, M_cut, M_1 = 10**logM, 10**logM_cut, 10**logM_1
    return (np.heaviside(M - kappa * M_cut, 0.5)*(M - kappa * M_cut) / M_1)**alpha * Nc

LRG_paper_params = {
    'logM_cut': logmcuts[..., None]*np.ones(hcos.ms.shape),  # 12.7
    'logM_1': logm1s[..., None]*np.ones(hcos.ms.shape),  # 13.6
    'sigma': sigmas[..., None]*np.ones(hcos.ms.shape),  # 0.2
    'alpha': alphas[..., None]*np.ones(hcos.ms.shape),  # 1.15
    'kappa': kappas[..., None]*np.ones(hcos.ms.shape),  # 0.08
    'f_ic': fics[..., None]*np.ones(hcos.ms.shape)  # .08
}

mthresh=10**LRG_paper_params['logM_cut'][:, 0]
hcos.add_custom_hod(name="LRG_HOD",mthresh=mthresh, ignore_existing=True, 
                    Nc_model=Nc_LRG, Ns_model=Ns_LRG, 
                    model_params=LRG_paper_params, z_dep=True)
```
